```python
df['GoodQT']=[1 if x >=7 else 0 for x in df['quality']]
df.sample(7)
```

```python
df.drop(['Id'],axis=1,inplace=True)
```

```python
X=df.drop(['GoodQT'],axis=1)
y=df['GoodQT']
```

```python
print(type(X))
print(type(y))
```

```python
df['GoodQT'].value_counts()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
def cls_eval(ytest,ypred):
    cm = confusion_matrix(ytest,ypred)
    print(cm)
    print('Accuracy Score',accuracy_score(ytest,ypred))
    print(classification_report(ytest,ypred))
    
def mscore(model):
    print('Training Score',model.score(X_train,y_train))  # Training Accuracy
    print('Testing Score',model.score(X_test,y_test))     # Testing Accuarcy
```

```python
log_reg = LogisticRegression(max_iter=1000,solver='liblinear')
log_reg.fit(X_train,y_train)
```

```python
mscore(log_reg)
```

```python
ypred_logreg=log_reg.predict(X_test)
```

```python

cls_eval(y_test,ypred_logreg)
```
